## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

````python
import matplotlib.pyplot as plt

# Plot Normal and StudentT
plt.figure()
pz.Normal(mu=0, sigma=1).plot_pdf(label='Normal(μ=0, σ=1)')
pz.Student(nu=5, mu=0, sigma=1).plot_pdf(support=(-4, 4), label='StudentT(ν=5, μ=0, σ=1)')
plt.show()

# Plot Exponential, Gamma, and HalfStudentT
plt.figure()
pz.Exponential(lam=1/2).plot_pdf(label='Exponential(λ=1/2)')
pz.Gamma(alpha=2, beta=1).plot_pdf(label='Gamma(α=2, β=1)')
pz.HalfStudentT(nu=5, sigma=1).plot_pdf(support=(0, 4), label='HalfStudentT(ν=5, σ=1)')
plt.show()
````
